```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
llm = ChatGroq(
    model="llama-3.3-70b-versatile",  # <-- Updated to a supported Llama 3.1 version
    temperature=0,
    groq_api_key=os.getenv("GROQ_API_KEY") 
)
```

```python
prompts= [
    {"role": "system", "content": "You are a helpful and polite AI programming assistant."},
    
    # Turn 1
    {"role": "user", "content": "Hello! My name is Sarthak."},
    {"role": "assistant", "content": "Namaste Sarthak! Great to meet you. How can I help you with your coding today?"},
    
    # Turn 2
    {"role": "user", "content": "What is my name?"},
    {"role": "assistant", "content": "Your name is Sarthak!"},
    
    # Turn 3 (Active Follow-up)
    {"role": "user", "content": "Can you explain what LangChain is in one sentence?"}
]

res = llm.invoke(prompts)
```

Manage History

```python
history = []
while True:
    query = input("user: ")
    
    if query.lower() in ["exit","bye","quit"]:
        print("Goodbye")
        break
    

    history.append( {"role": "user", "content": query})
    print("User: ",query)
    
    res = llm.invoke(history)
    history.append( {"role": "ai", "content": res.content})
    
    print("AI: ",res.content,"\n")
```
