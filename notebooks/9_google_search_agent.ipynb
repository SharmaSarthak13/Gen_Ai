```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_community.utilities import GoogleSerperAPIWrapper
import os
from langchain_groq import ChatGroq
from langchain.agents import create_agent
```

```python
llm = ChatGroq(
    model="llama-3.3-70b-versatile",  # <-- Updated to a supported Llama 3.1 version
    temperature=0,
    groq_api_key=os.getenv("GROQ_API_KEY") 
)
search = GoogleSerperAPIWrapper()
```

```python
agent = create_agent(
    model = llm,
    tools = [search.run],
    system_prompt = "You are a agent and can search for any question on google "
)
```

```python
question = "where is baghpat?"
```

```python
res = agent.invoke({"messages":[{"role":"user","content":question}]})
```

```python
print(res["messages"][-1].content)
```
