```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import os 

from langchain_groq import ChatGroq
llm = ChatGroq(
    model="llama-3.3-70b-versatile", 
    streaming = True,
    temperature=0,
    groq_api_key=os.getenv("GROQ_API_KEY") 
)
```

```python
ques = "what is gen ai ?"
```

```python
res = llm.stream(ques)
for chunk in res:
    print(chunk.content,end = "")
```
