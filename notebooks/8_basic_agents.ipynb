```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add_numbers(a:int , b:int):
    """
       it will return the sum of two numbers
       
       Args:
            a:Number One
            b:Number Two
    """
    
    return a+b

@tool
def multiply_numbers(a:int , b:int):
    """
       it will return the product of two numbers
       
       Args:
            a:Number One
            b:Number Two
    """
    
    return a*b
```

```python
import os 

from langchain_groq import ChatGroq
from langchain.agents import create_agent
```

```python
llm = ChatGroq(
    model="llama-3.3-70b-versatile", 
    streaming = True,
    temperature=0,
    groq_api_key=os.getenv("GROQ_API_KEY") 
)
```

```python
agent = create_agent(
    model = llm,
    tools = [add_numbers,multiply_numbers],
    system_prompt = "You are a math teacher , and always ise tool for calculation"
)
```

```python
response = agent.invoke({"messages":[{"role":"user","content":"what is 2+30*78?" }]})
```
