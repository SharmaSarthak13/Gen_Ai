```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import os 

from langchain_groq import ChatGroq
llm = ChatGroq(
    model="llama-3.3-70b-versatile",  # <-- Updated to a supported Llama 3.1 version
    temperature=0,
    groq_api_key=os.getenv("GROQ_API_KEY") 
)
```

### Discuss the problem first

```python
text = "Hello , my name is Sarthak ,"\
    "email is isarthak2005@gmail.com ,"\
    "and age is 20"
```

```python
res = llm.invoke(f"please give me only name , email and age from this text{text}")
```

```python
res.content
```

```python
print(type(res.content))
```

### Solution is Structured Output (with help of pydantic)
#### machine to machine communication in same way 


```python
from pydantic import BaseModel , Field
from typing import List
class ResponseStructure(BaseModel):
    name:str = Field(description = "Complete Name ")
    age:int = Field(description = "Email Address ")
    email:str = Field(description = "Age ")
    
structured_llm = llm.with_structured_output(ResponseStructure)
```

```python
res = structured_llm.invoke(f"please give me only name , email and age from this text{text}")
```

```python
res
```

```python
print(res.age)
print(res.name)
print(res.email)
```

```python
res.model_dump()
```

```python
class Movies(BaseModel):
    title:str = Field(description = "Movie Title")
    year: int = Field(description="Movie Release Year")  
    
    
class AllMovies(BaseModel):
    movies:List[Movies]
    
structured_llm_movie = llm.with_structured_output(AllMovies)

```

```python
structured_llm_movie.invoke("gave me one recent movie")
```

```python
structured_llm_movie.invoke("gave me three romantic movie")
```
