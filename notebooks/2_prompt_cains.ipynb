```python
import os
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser



```

```python
# Initialize the Groq LLM with a supported model
llm_groq = ChatGroq(
    model="llama-3.3-70b-versatile",  # <-- Updated to a supported Llama 3.1 version
    temperature=0,
    groq_api_key=os.getenv("GROQ_API_KEY") 
)
out = StrOutputParser()
```

```python
def transform_case(result:str):
    return result.upper()
```

```python
prompts = ChatPromptTemplate.from_messages([
    {"role": "system", "content": "you are a translator, translate in {language}"},
    {"role": "user", "content": "{query}"}
])
```

```python
chains = prompts | llm_groq | out | transform_case

res = chains.invoke({
    "language": "hinglish",
    "query": "hello my name is sarthak"
})

print(res)
```
