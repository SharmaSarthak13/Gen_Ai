```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
texts = "What this error actually means is: You asked me to parse a JSON response, but what I actually got back was blank, an HTML error page, or plain text. Because it didn't find a { or [ at the very beginning (character 0), Python's JSON decoder threw its hands up."
```

```python
spilitter = RecursiveCharacterTextSplitter(chunk_size = 50,chunk_overlap=20)
```

```python
chunks = spilitter.split_text(texts)
```
