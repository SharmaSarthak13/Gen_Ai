```python
import os
from langchain_groq import ChatGroq
from langchain_huggingface import HuggingFaceEmbeddings

# 1. Initialize your Groq LLM for text generation/reasoning
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0,
    groq_api_key=os.environ.get("GROQ_API_KEY")
)

# 2. Initialize a free Hugging Face model for creating vector embeddings
# (This model runs locally on your machine and is great for text retrieval)
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")

# --- How to use them together ---

text_to_embed = "Generative AI is transforming software development."

# This creates your numerical vector (list of floats)
vector = embeddings.embed_query(text_to_embed)
print(f"Embedding Vector Length: {len(vector)}")  # Usually 768 dimensions

# If you want to pass text to your LLM later, you use 'llm' normally:
# response = llm.invoke(f"Summarize this concept: {text_to_embed}")
```
